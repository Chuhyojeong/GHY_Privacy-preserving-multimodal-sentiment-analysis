```python
import pickle
import matplotlib.pyplot as plt
from PIL import Image

# .pickle 파일 로드
with open('./dataset/mosi/raw/video.pickle', 'rb') as f:
    image_data = pickle.load(f, encoding='latin1')

# 이미지 전처리
processed_image = []

# 이미지 데이터 형태 통일
for image in image_data:
    # 이미지 크기 조정
    resized_image = Image.fromarray(image).resize((224, 224))
    processed_image.append(resized_image)

# 이미지 전처리 및 부동 소수점 변환
processed_image = [np.array(image, dtype=np.float32) / 255.0 for image in processed_image]

# 이미지 출력
plt.imshow(processed_image[0])
plt.axis('off')
plt.show()


# .pickle 파일 로드
with open('./dataset/mosi/raw/video_2way.pickle', 'rb') as f:
    image_data = pickle.load(f, encoding='latin1')

# 이미지 전처리
processed_image = []

# 이미지 데이터 형태 통일
for image in image_data:
    # 이미지 크기 조정
    resized_image = Image.fromarray(image).resize((224, 224))
    processed_image.append(resized_image)

# 이미지 전처리 및 부동 소수점 변환
processed_image = [np.array(image, dtype=np.float32) / 255.0 for image in processed_image]

# 이미지 출력
plt.imshow(processed_image[0])
plt.axis('off')
plt.show()
```

```python
import pickle
import matplotlib.pyplot as plt

# .pickle 파일 로드
with open('./dataset/mosi/raw/video.pickle', 'rb') as f:
    image_data = pickle.load(f, encoding='latin1')

# 이미지 데이터 하나만 선택
image = image_data[0]  # 첫 번째 이미지 데이터 선택

# 각 채널 시각화
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(image[:,:,0], cmap='Reds')
axes[0].set_title('Red Channel')
axes[0].axis('off')
axes[1].imshow(image[:,:,1], cmap='Greens')
axes[1].set_title('Green Channel')
axes[1].axis('off')
axes[2].imshow(image[:,:,2], cmap='Blues')
axes[2].set_title('Blue Channel')
axes[2].axis('off')

plt.show()
```
